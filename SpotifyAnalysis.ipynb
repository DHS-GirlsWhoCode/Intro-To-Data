# Intro to Data with Spotify

**Part One:** Load Data Into Pandas DataFrame

```python
import pandas as pd
import seaborn as sns
```

```python
filepath = "https://raw.githubusercontent.com/DHS-GirlsWhoCode/Intro-To-Data/main/spotify.csv"
sptfy = pd.read_csv(filepath, index_col="Date", parse_dates=True)
```

**Part Three:** Visualization

```python
sns.lineplot(data=sptfy)
```

```python
# can index at column data to graph individual line plots
sns.lineplot(data=sptfy['Something Just Like This'])
```
